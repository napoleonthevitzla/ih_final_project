```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Exportamos datasets(elegir esta opción )

#Data from berseker source
fisio_raw = pd.read_csv('./data/DataSet_Berseker_Fisio_session.csv')
mountain_raw = pd.read_csv('./data/DataSet_Berseker_Mountain_session.csv')
```

```python
# Mountain manipulate

mountain_raw['Category_fitness'] = "Mountain"
# En running tengo que trabajar con Marca temporal en lugar de la columna día
mountain_raw = mountain_raw.drop(['Marca temporal'], axis=1)
mountain_raw['date'] = pd.to_datetime(mountain_raw['Dia'], format="%d/%m/%Y", errors='coerce')
mountain_raw['year']= mountain_raw['date'].dt.year
mountain_raw['month']= mountain_raw['date'].dt.month
mountain_raw['day']= mountain_raw['date'].dt.day
mountain_raw['week_of_year'] = mountain_raw['date'].dt.week
mountain_raw['day_of_week'] = mountain_raw['date'].dt.dayofweek
mountain_raw['is_leap_year'] = mountain_raw['date'].dt.is_leap_year
    
dw_mapping={
    0: 'Monday', 
    1: 'Tuesday', 
    2: 'Wednesday', 
    3: 'Thursday', 
    4: 'Friday',
    5: 'Saturday', 
    6: 'Sunday'
    } 
mountain_raw['day_of_week_name']=mountain_raw['date'].dt.weekday.map(dw_mapping)

mountain_raw.head()
```

```python
# FISIO_RAW Data manipulation
fisio_raw['Category_fitness'] = "Fisio"
# En running tengo que trabajar con Marca temporal en lugar de la columna día
fisio_raw = fisio_raw.drop(['Marca temporal'], axis=1)
fisio_raw['date'] = pd.to_datetime(fisio_raw['Dia'], format="%d/%m/%Y", errors='coerce')
fisio_raw['year']= fisio_raw['date'].dt.year
fisio_raw['month']= fisio_raw['date'].dt.month
fisio_raw['day']= fisio_raw['date'].dt.day
fisio_raw['week_of_year'] = fisio_raw['date'].dt.week
fisio_raw['day_of_week'] = fisio_raw['date'].dt.dayofweek
fisio_raw['is_leap_year'] = fisio_raw['date'].dt.is_leap_year
    
fisio_raw['day_of_week_name']=fisio_raw['date'].dt.weekday.map(dw_mapping)
fisio_raw.head()
```
