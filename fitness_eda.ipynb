```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Exportamos datasets(elegir esta opción )

#Data from berseker source
running_raw = pd.read_csv('./data/DataSet_Berseker2022_running_session.csv')
fisio_raw = pd.read_csv('./data/DataSet_Berseker_Fisio_session.csv')
mountain_raw = pd.read_csv('./data/DataSet_Berseker_Mountain_session.csv')

```

```python
# Fisio / mountain / running datasets : CASO FISIO
#fisio_raw.head()
#fisio_raw.describe()
#fisio_raw.dtypes
running_raw['date'] = pd.to_datetime(running_raw['Dia'], format="%d/%m/%Y %H:%M:%S", errors='coerce')

```

```python
# FISIO_RAW Data manipulation


#Incluir columna con categoria 
fisio_raw['Category_fitness'] = "Fisio"

# Eliminar marca temporal
fisio_raw = fisio_raw.drop(['Marca temporal'], axis=1)

#Incluir semana del año y mes del año y el día de la semana

#Convert to datetime format:
fisio_raw['date'] = pd.to_datetime(fisio_raw['Dia'], format="%d/%m/%Y", errors='coerce')
# Get year, month and day
fisio_raw['year']= fisio_raw['date'].dt.year
fisio_raw['month']= fisio_raw['date'].dt.month
fisio_raw['day']= fisio_raw['date'].dt.day
# Get the week of year, the day of week and leap year
fisio_raw['week_of_year'] = fisio_raw['date'].dt.week
fisio_raw['day_of_week'] = fisio_raw['date'].dt.dayofweek
fisio_raw['is_leap_year'] = fisio_raw['date'].dt.is_leap_year

#  Map to get the day of the week 
dw_mapping={
    0: 'Monday', 
    1: 'Tuesday', 
    2: 'Wednesday', 
    3: 'Thursday', 
    4: 'Friday',
    5: 'Saturday', 
    6: 'Sunday'
} 
fisio_raw['day_of_week_name']=fisio_raw['date'].dt.weekday.map(dw_mapping)

# Incluir minutos de sesión en el dataset de Fisio con valores aleatorios entre 60 y 90 minutos

fisio_raw['Min_sesion'] = np.random.randint(60, 90, size = len(fisio_raw))
#fisio_raw['Min_sesion'] = np.random.randint(60, 90, fisio_raw.shape[0])
fisio_raw.head()
```

```python
# FISIO_RAW Data manipulation

def databerseker (dataset, activity):
    dataset['Category_fitness'] = activity
    dataset = dataset.drop(['Marca temporal'], axis=1)
    dataset['date'] = pd.to_datetime(dataset['Dia'], format="%d/%m/%Y", errors='coerce')
    dataset['year']= dataset['date'].dt.year
    dataset['month']= dataset['date'].dt.month
    dataset['day']= dataset['date'].dt.day
    dataset['week_of_year'] = dataset['date'].dt.week
    dataset['day_of_week'] = dataset['date'].dt.dayofweek
    dataset['is_leap_year'] = dataset['date'].dt.is_leap_year
    
    dw_mapping={
    0: 'Monday', 
    1: 'Tuesday', 
    2: 'Wednesday', 
    3: 'Thursday', 
    4: 'Friday',
    5: 'Saturday', 
    6: 'Sunday'
    } 
    dataset['day_of_week_name']=dataset['date'].dt.weekday.map(dw_mapping)
    
```
