```python
#Incluir columna con categoria 
fisio_raw['Category_fitness'] = "Fisio"
```

```python
# Eliminar marca temporal
fisio_raw = fisio_raw.drop(['Marca temporal'], axis=1)
```

```python
#Incluir semana del año y mes del año

#Convert to datetime format:
fisio_raw['date'] = pd.to_datetime(fisio_raw['Dia'], format="%d/%m/%Y", errors='coerce')
# Get year, month and day
fisio_raw['year']= fisio_raw['date'].dt.year
fisio_raw['month']= fisio_raw['date'].dt.month
fisio_raw['day']= fisio_raw['date'].dt.day
# Get the week of year, the day of week and leap year
fisio_raw['week_of_year'] = fisio_raw['date'].dt.week
fisio_raw['day_of_week'] = fisio_raw['date'].dt.dayofweek
fisio_raw['is_leap_year'] = fisio_raw['date'].dt.is_leap_year

#  Map to get the day of the week 
dw_mapping={
    0: 'Monday', 
    1: 'Tuesday', 
    2: 'Wednesday', 
    3: 'Thursday', 
    4: 'Friday',
    5: 'Saturday', 
    6: 'Sunday'
} 
fisio_raw['day_of_week_name']=fisio_raw['date'].dt.weekday.map(dw_mapping)


fisio_raw.head(100)
```
