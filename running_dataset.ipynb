```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Exportamos datasets(elegir esta opción )

#Data from berseker source
running_raw = pd.read_csv('./data/DataSet_Berseker2022_running_session.csv')
fisio_raw = pd.read_csv('./data/DataSet_Berseker_Fisio_session.csv')
mountain_raw = pd.read_csv('./data/DataSet_Berseker_Mountain_session.csv')
```

```python
running_raw['Category_fitness'] = "Running"
running_raw['Dia'] = pd.to_datetime(running_raw['Marca temporal'], format="%d/%m/%Y %H:%M:%S", errors='coerce')
running_raw['date'] = pd.to_datetime(running_raw['Dia'], format="%Y-%m-%d %H:%M:%S", errors='coerce')
running_raw = running_raw.drop(['Marca temporal'], axis=1)
running_raw['year']= running_raw['date'].dt.year
running_raw['month']= running_raw['date'].dt.month
running_raw['day']= running_raw['date'].dt.day
running_raw['week_of_year'] = running_raw['date'].dt.week
running_raw['day_of_week'] = running_raw['date'].dt.dayofweek
running_raw['is_leap_year'] = running_raw['date'].dt.is_leap_year
dw_mapping={
    0: 'Monday', 
    1: 'Tuesday', 
    2: 'Wednesday', 
    3: 'Thursday', 
    4: 'Friday',
    5: 'Saturday', 
    6: 'Sunday'
    } 

running_raw['day_of_week_name']=running_raw['date'].dt.weekday.map(dw_mapping)

running_raw.head()

```
